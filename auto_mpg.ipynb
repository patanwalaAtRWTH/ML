### The test train split

```python
d1.origin = d1.origin.replace({
    'America':1,
    'Europe':2,
    'Asia':3
})
```

```python
np.isnan(d1).any()
```

```python
np.isinf(d1).any()
```

```python
d1.isna().sum()
```

```python
d1.shape
```

```python
d1.dropna(axis=0, inplace=True)
```

```python
d1.shape
```

```python
label_column = 'mpg'
feature_column = [c for c in d1.columns if c!= label_column]
feature_column.remove('log_weight')
feature_column.remove('double_mpg')
feature_column.remove('hp_per_lb')
X = d1[feature_column].values
y = d1[label_column]
```

```python
X.shape, y.shape
```

```python
feature_column
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                   test_size=0.25, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

lin_model = make_pipeline(StandardScaler(),
                         LinearRegression()).fit(X_train, y_train)
```

```python
print(f'Intercept:{lin_model[1].intercept_:.4f}')
for i, coef in enumerate(lin_model[1].coef_):
    print(f"Coefficient for {feature_column[i]}:{coef:.4f}")
```

```python
from sklearn.metrics import mean_squared_error

test_prediction = lin_model.predict(X_test)
lin_model_mse = mean_squared_error(y_test, test_prediction)
```

```python
lin_model_mse
```

### Quadratic Regression

```python
from sklearn.preprocessing import PolynomialFeatures

poly_model = make_pipeline(StandardScaler(),
                          PolynomialFeatures(degree=2),
                          LinearRegression()
                          ).fit(X_train, y_train)
```

```python
poly_model[2].intercept_.round(4)
```

```python
poly_model[2].coef_
```

```python
print(f'The quad model has {len(poly_model[2].coef_)+1} parameters')
```

```python
test_predictions_quad = poly_model.predict(X_test)
poly_model_mse = mean_squared_error(y_test, test_predictions_quad)
```

```python
print(f"Quad regression model Test MSE: {poly_model_mse:.4f}")
```

### Quad + LASSO

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold


kfold = KFold(n_splits=5, shuffle=True,
             random_state=0)

poly_lasso = make_pipeline(StandardScaler(),
                          PolynomialFeatures(degree=2),
                          GridSearchCV(estimator=
                                      Lasso(max_iter=10000),
                                      param_grid={
                                          'alpha':np.logspace(-3,0,25)
                                      },
                                      scoring='neg_mean_squared_error',
                                      iid=False,
                                      cv=kfold)
                          ).fit(X_train, y_train)
```

```python
poly_lasso[2].best_params_
```

```python
print(f"Intercept: {poly_lasso[2].best_estimator_.intercept_:.4f}")
for i, coef in enumerate(poly_lasso[2].best_estimator_.coef_):
    print(f"Coefficients #{i:2}: {coef:7.4f}")
```

```python
test_prediction_quad_lasso = poly_lasso.predict(X_test)
polylasso_mse = mean_squared_error(y_test, test_prediction_quad_lasso)
```

```python
print(f"Quad regression + LASSO model. Test MSE: {polylasso_mse:.4f}")
```

```python
print(f"Linear Regression : {lin_model_mse}")
print(f"Quad Regression: {poly_model_mse}")
print(f"Poly_LASSO: {polylasso_mse}")
```

```python
from sklearn.metrics import r2_score

print(f"Linear Regression r2: {r2_score(y_test, lin_model.predict(X_test)):.4f}")
print(f"Quad Regression r2: {r2_score(y_test, poly_model.predict(X_test)):.4f}")
print(f"Poly_LASSO r2: {r2_score(y_test, poly_lasso.predict(X_test)):.4f}")
```

### Final Model

```python
final_lasso = poly_lasso[2].best_estimator_
final_model = make_pipeline(StandardScaler(),
                           PolynomialFeatures(degree=2),
                           final_lasso
                           ).fit(X,y)
```
